```python
import pandas as pd

df = pd.read_csv("results/1_0.csv", index_col=0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,5))
sns.violinplot(
    data=df,
    x='model_name', 
    y='mean_fit_time', 
    ax=ax
)
ax.set_ylim(0, 1)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.violinplot(
    data=df,
    x='model_name', 
    y='mean_test_neg_root_mean_squared_error', 
    ax=ax
)
ax.set_ylim(-100,100)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.violinplot(
    data=df,
    x='model_name', 
    y='mean_test_r2', 
    ax=ax
)
```

```python
import ast

avg_moddescribe2 = df.groupby('model_name').mean_test_r2.mean().sort_values()
display(avg_moddescribe2)



for model_name in avg_moddescribe2.index: 
    par_df = df[df.model_name==model_name].reset_index()
    par_params = pd.DataFrame(par_df.params.apply(lambda x: ast.literal_eval(x)).to_list())
    par_params = par_params.fillna(-1)
    par_df = par_df.join(par_params)
    stats = []
    for param_name in par_params.columns:
        fig, ax = plt.subplots(figsize=(12,5))
        sns.violinplot(
            data=par_df,
            x=param_name, 
            y='mean_test_r2', 
            ax=ax,
        )
        ax.set_title(model_name)
        plt.show()
        display(par_df.groupby(param_name).mean_test_r2.describe())
```

```python
df.sort_values('mean_test_r2', ascending=False).head(10).reset_index().loc[0].params
```

```python
for target in df.target.unique():
    head = df[df.target==target].sort_values('mean_test_r2', ascending=False).head(3)
    print(head.iloc[0].params)
    print(head.iloc[1].params)
    print(head.iloc[2].params)
    display(head)
```

```python
df_good_models = df[df.mean_test_r2>0.60]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.set_palette("Paired")
sns.swarmplot(data=df_good_models, x='target', y='mean_test_r2', hue='model_name', ax=ax)

plt.title('Scatterplot of target vs. mean_test_r2 by model_name')
plt.xlabel('target')
plt.ylabel('mean_test_r2')

plt.show()
```
