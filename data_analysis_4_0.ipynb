```python
import pandas as pd

df = pd.read_csv("results/4_0_final.csv", index_col=0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_good_models = df[df.mean_test_r2>0.60]

fig, ax = plt.subplots(figsize=(12, 8))

sns.set_palette("Paired")
sns.swarmplot(data=df_good_models, x='target', y='mean_test_r2', hue='model_name', ax=ax)

plt.title('Scatterplot of target vs. mean_test_r2 by model_name')
plt.xlabel('target')
plt.ylabel('mean_test_r2')

plt.show()
```
